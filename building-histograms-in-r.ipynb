```r
set.seed(123)
data1 <- runif(100, min = 0, max = 50)

hist(data1,
     breaks = 10,
     main = "Histogram of Random Data",
     xlab = "Value",
     ylab = "Frequency",
     col = "lightpink",
     border = "black")
```

```r
# Sample Monthly Sales Data (12 months)
months <- month.name 
set.seed(789)  
sales <- round(runif(12, min = 1000, max = 10000))  # Random sales between 1,000 and 10,000

monthly_sales <- data.frame(Month = months, Sales = sales)

print(monthly_sales)

write.csv(monthly_sales, "monthly_sales.csv", row.names = FALSE)

cat("File 'monthly_sales.csv' created in current working directory.")

sales_data <- read.csv("monthly_sales.csv")


hist(sales_data$Sales,
     col = "purple",
     main = "Company Sales Data of Products",
     xlab = "Product Sales",
     ylab = "Frequency")

legend("topright",
       legend = "Sales Histogram",
       fill = "purple",
       border = "black")
```

```r
set.seed(123) 
exam_scores <- sample(40:100, 50, replace = TRUE)

bar_colors <- ifelse(exam_scores >= 60, "maroon", "orange")

hist(exam_scores,
     breaks = 8,
     col = bar_colors,
     border = "black",
     main = "Histogram of Exam Scores",
     xlab = "Scores",
     ylab = "Number of Students")
```
