```r
categories <- c("A", "B", "C", "D", "E")
values <- c(15, 30, 45, 20, 10)
barplot(values, names.arg = categories,
        main = "Customized Bar Plot",
        xlab = "Categories",
        ylab = "Values",
        col = "blue",
        border = "black")
```

```r
barplot(values,
        names.arg = categories,
        main = "Customized Bar Plot",
        xlab = "Categories",
        ylab = "Values",
        col = c("blue", "green", "red", "purple", "orange"),
        border = "black",
        ylim = c(0, max(values) + 10),
        width = 0.5,
        beside = TRUE)
```

```r
data <- data.frame(
  categories = c("A", "B", "C", "D"),
  group1 = c(10, 15, 7, 12),
  group2 = c(8, 11, 5, 9)
)

barplot(
  height = t(as.matrix(data[, c("group1", "group2")])),
  beside = TRUE,
  col = c("blue", "green"),
  names.arg = data$categories,
  legend.text = c("Group 1", "Group 2"),
  xlab = "Categories",
  ylab = "Values",
  main = "Side-by-Side Bar Chart"
)
```

```r
data <- data.frame(
  name = c("Wini","Manu","Mia"),
  fiction = c(5, 3, 6),
  nonfiction = c(2, 4, 1)
)

bar_matrix <- t(as.matrix(data[, c("fiction", "nonfiction")]))
colnames(bar_matrix) <- data$name

barplot(
  bar_matrix,
  beside = FALSE,
  col = c("skyblue", "salmon"),
  legend.text = c("Fiction", "Nonfiction"),
  args.legend = list(x = "topright", title = "Genre"),
  main = "Books Read by Genre",
  ylab = "Number of Books",
  xlab = "Individuals"
)
```

```r
data(mtcars)

avg_mpg <- tapply(mtcars$mpg, mtcars$cyl, mean)

barplot(
  avg_mpg,
  horiz = TRUE,
  col = "green",
  main = "Average MPG by Cylinder",
  xlab = "MPG",
  ylab = "Number of Cylinders",
  names.arg = names(avg_mpg),
  las = 1 
)

text(avg_mpg,
     y = barplot(avg_mpg, horiz = TRUE, plot = FALSE),
     labels = round(avg_mpg, 1),
     pos = 4, col = "black"
)
```
